```python
scoring = 'r2'
results_bls = np.zeros_like(per_change)
results_ls = np.zeros_like(per_change)
results_pls = np.zeros_like(per_change)
gerror = np.zeros_like(per_change)
r = np.zeros_like(per_change)
rerror = np.zeros_like(per_change)
err_var = np.zeros_like(per_change)
alphas = np.logspace(-4,4,100)
Ks = []
Kstg = []
w = []
wg = []
pcases = []
pscores = []
s = []
```

```python
for j in np.arange(len(subsc)):
    test_id = subsc[j]
    test_index = subsc == test_id
    train_index = subsc != test_id
    X_train = X_all_c[train_index,:,:]
    X_test = X_all_c[test_index,:,:]
    y_train = per_change[train_index]
    y_test = per_change[test_index]
    X0_ss = np.zeros((X_train.shape[0],2*len(cr_feats)+1))
    X_test_ss = np.zeros((1,2*len(cr_feats)+1))
    idy = y_train[y_train<=0.3]
    

    # Cross validation
    cvn = 6
    X0_ss0,scaler_ss,X_test_ss0 = util.model_scale(skp.StandardScaler(),
                                                X_train,train_index,X_test,test_index,pre_updrs_off,False,False)
    Kc = K.reshape(1, -1)
    filtered = fnmatch.filter(Kc.ravel(), 'gradient*')

    
    X0_ss = np.zeros((X_train.shape[0],2*len(cr_feats)+len(filtered)+1))
    X_test_ss = np.zeros((1,2*len(cr_feats)+len(filtered)+1))
    
    pc = 0
    for p in np.arange(X0_ss0.shape[1]):
        if Kc[0][p] in cr_feats or Kc[0][p] in filtered:
            X0_ss[:,pc] = X0_ss0[:,p]
            X_test_ss[0,pc] = X_test_ss0[0,p]
            pc = pc+1
            lr = stats.linregress(X0_ss0[:,p],y_train)
            #print(lr.rvalue)


    with np.errstate(divide='ignore', invalid='ignore'):
      # Feature selection
      print('Overwriting clinically relevant features')
      sel = skf.SelectKBest(skf.r_regression,k=200)
      X0_ss = sel.fit_transform(X0_ss0,y_train)
      X_test_ss = sel.transform(X_test_ss0)
      Ks.append(sel.transform(K.reshape(1, -1)))
    dx, y_n = cKDTree(X0_ss).query(X_test_ss, k=1)

    # LASSO
    with warnings.catch_warnings():
      warnings.filterwarnings("ignore", category=ConvergenceWarning)
      lasso = skn.KNeighborsRegressor(n_neighbors=3,weights='distance')#slm.RidgeCV()
      est_ls = lasso.fit(X0_ss,y_train)
    results_ls[j] = est_ls.predict(X_test_ss)

    # # Partial least squares
    pls = skd.PLSRegression(n_components=10)
    est_pls = pls.fit(X0_ss,y_train)
    results_pls[j] = est_pls.predict(X_test_ss)
    print('Lasso predicts',str(np.round(results_ls[j],2)),
              'for case with',str(np.round(per_change[j],2)),
              'and neighbor',str(np.round(y_train[y_n],2).item()))
    gerror[j] = (abs(results_ls[j]-y_test))
        
    if gerror[j] > 0.15:
          pcases.append(test_id)
          pscores.append(s)
  
```
